```python
import numpy as np
import pandas as pd
```

```python
df_origin = pd.read_csv('./Data_Lake/Bronze/cars-origem.csv',sep=';')
```

```python
df_origin
```

```python
#Tornar todas as letras minúsculas
df_origin['brand'] = df_origin['brand'].str.lower()
```

```python
#Retirar todos os espaços
df_origin['brand'] = df_origin['brand'].str.replace(" ","")
```

```python
df_origin
```

```python
df_cars = pd.read_csv('./Data_Lake/Bronze/cars.csv',sep=';')
```

```python
df_cars
```

```python
#Unificar nome das colunas: make = brand
df_cars.rename(columns = {'make':'brand'},inplace=True)
```

```python
df_cars['price'].head(20)
```

```python
df_cars_new = df_cars.merge(df_origin)   #alternativa: join!
```

```python
df_cars_new
```

```python
df_cars_new.dtypes
```

```python
#Primeiro, descobrimos que os preços estão em formato "objetct" e não "float"
df_cars_new.loc[185,'price']
```

```python
#Vamos eliminar todas as linhas que tem ? no PRICE, HORSEPOWER E PEAK-RPM
df_cars_new = df_cars_new[df_cars_new['price'] != '?']
df_cars_new = df_cars_new[df_cars_new['horsepower'] != '?']
df_cars_new = df_cars_new[df_cars_new['peak-rpm'] != '?']

df_cars_new
#df_cars_new['price'] = df_cars_new['price'].replace("?","-1")
```

```python
#Vamos modificar o formato do price, horsepower and peak-rpm para serem números
df_cars_new = df_cars_new.astype(dtype={"price": np.float64,"horsepower":np.float64,"peak-rpm":np.float64})
df_cars_new.dtypes
```

```python
#f_cars_new.head(15)
```

```python
byComp = df_cars_new.groupby('brand')
```

```python
#Tarefa 1: Achar preço médio das marcas europeias
df_cars_europe = df_cars_new.loc[df_cars_new['continent']=='Europe']
byComp_Europe = df_cars_europe.groupby('brand')
byComp_Europe.mean()['price']
```

```python
#Tarefa 2: Achar marca do carro mais comprado de cada continente
byContinent = df_cars_new.groupby('continent')
byContinent
```

```python
df_cars_new.groupby(['brand','continent']).mean()
```

```python
df_cars_new[['brand','continent']].value_counts()
```

```python
#Vamos transformar a série em data frame
df_brand_continent = df_cars_new[['brand','continent']].value_counts().to_frame()
df_brand_continent
```

```python
#Vamos renomear a nova coluna que se chama 0 para se chamar "Quantity"
df_brand_continent.rename(columns = {0:'quantity'},inplace=True)
df_brand_continent
```

```python
#Vamos transformar os multi-índices em colunas
df_brand_continent.reset_index(level=['brand', 'continent'],inplace=True)
```
